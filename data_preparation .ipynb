```python
import pandas as pd
```

```python
df1=pd.read_csv('Apple.csv')
df1.head()
```

```python
date=df1['Date']
close=df1['Close']
```

```python
#creat new column of the same size 
change=[0]*len(date)
momentum=[0]*len(date)
```

```python
#calculate momentum and change
for i in range(1,len(date)):
    if close[i]>close[i-1]:
        momentum[i]='1'
        change[i]=(close[i]-close[i-1])/close[i-1]
    else:
        momentum[i]='0'
        change[i]=(close[i-1]-close[i])/close[i-1]
```

```python
df1['Change']=change
df1['Momentum']=momentum
df1.head()
```

```python
#save important parameters
x1=pd.DataFrame({'Date':date, 'Close':close,'Change':change,'Momentum':momentum})
x1.to_csv("Apple_Modified.csv",index=False,header=True)
```
